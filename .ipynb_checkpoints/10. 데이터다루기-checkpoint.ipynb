# 10.1 데이터 탐색

```python
from typing import List,Dict
from collections import Counter
import math

import matplotlib.pyplot as plt
import random
from probability import inverse_normal_cdf
from linear_algebra import shape, get_row, get_column, make_matrix, vector_mean, vector_sum, dot, magnitude, vector_subtract, scalar_multiply
```

```python
def bucketsize(point: float,bucket_size: float)->float:
    return bucket_size*math.floor(point/bucket_size)

def make_histogram(points:List[float],bucket_size:float)->Dict[float,int]:
    return Counter(bucketsize(point,bucket_size) for point in points)

def plot_histogram(points:List[float], bucket_size:float,title:str=""):
    histogram=make_histogram(points,bucket_size)
    plt.bar(histogram.keys(),histogram.values(),width=bucket_size)
    plt.title(title)
```

```python
random.seed(0)

uniform=[200*random.random() - 100 for _ in range(10000)]

normal=[57*inverse_normal_cdf(random.random()) for _ in range(10000)]
```

```python
plot_histogram(uniform,10,"Uniform Histogram") # 유니폼
```

```python
plot_histogram(normal,10,"Normal Histogram") # 정규분포
```

```python
# 2차원 데이터
def random_normal()->float:
    return inverse_normal_cdf(random.random())

xs=[random_normal() for _ in range(1000)]
ys1=[x+random_normal()/2 for x in xs]
ys2=[-x+random_normal() /2 for x in xs]
```

```python
plot_histogram(ys1,10)
```

```python
plot_histogram(ys2,10)
```

```python
#결합확률분포
plt.scatter(xs,ys1,marker='.',color='red',label='ys1')
plt.scatter(xs,ys2,marker='.',color='blue',label='ys2')
plt.xlabel('xs')
plt.ylabel('ys')
plt.legend(loc=9)
plt.title("Very Different Joint Distributions")
plt.show()
```

```python
#상관관계로 차이 확인
from statistics import correlation

print(correlation(xs,ys1))
print(correlation(xs,ys2))
```

```python
# 다차원 데이터의 관계 상관관계
from linear_algebra import Matrix,Vector,make_matrix
```

```python
def correlation_matrix(data:List[Vector])->Matrix:
    def correlation_ij(i:int,j:int)->float:
        return correlation(data[i],data[j])
    
    return make_matrix(len(data),len(data),correlation_ij)
```

```python
dimen=100
def random_row() -> List:
    row=[None,None,None,None]
    row[0]=random_normal()
    row[1]=-5 *row[0]+random_normal()
    row[2]=row[0]+row[1]+5*random_normal()
    row[3]=6 if row[2] >-2 else 0
    return row

corr_data=[random_row() for _ in range(dimen)]
print(corr_data)
```

```python
_,num_vectors=shape(corr_data)
fig,ax=plt.subplots(num_vectors,num_vectors)

for i in range(num_vectors):
    for j in range(num_vectors):
        
        if i != j: ax[i][j].scatter(get_column(corr_data,j),get_column(corr_data,i))
        
        else: ax[i][j].annotate("series"+str(i),(0.5,0.5), xycoords='axes fraction', ha="center", va="center")
        
        if i<num_vectors -1: ax[i][j].xaxis.set_visible(False)
        if j>0: ax[i][j].yaxis.set_visible(False)
            

ax[-1][-1].set_xlim(ax[0][-1].get_xlim())
ax[0][0].set_ylim(ax[0][1].get_ylim())
plt.show()
```

# 10.4 데이터 정제하고 합치기

```python
from dateutil.parser import parse
```

```python
def parse_row(row:List[str])->StockPrice:
    symbol,date,closing_price=row
    return StockPrice(symbol=symbol,date=parse(date).date(),closing_price=float(closing_price))

stock=parse_row(["MSFT","2018-12-14","106.03"])
print(stock)
```

```python
# improve
from typing import Optional
import re

def try_parse_row(row:List[str])->Optional[StockPrice]:
    symbol,date_,closing_price_=row
    
    if not re.match(r"^[A-Z]+$",symbol):
        return None
    try:
        date=parse(date_).date()
    except ValueError:
        return None
    
    try:
        closing_price=float(closing_price_)
    except ValueError:
        return None
    
    return StockPrice(symbol,date,closing_price)

try_stock=try_parse_row(["MSFT","2018-12-14","106.03"])
print(try_stock)
try_stock=try_parse_row(["MSFT0","2018-12-14","106.03"])
print(try_stock)
try_stock=try_parse_row(["MSFT","2018-12-14","x"])
print(try_stock)
try_stock=try_parse_row(["MSFT","2018-12--14","106.03"])
print(try_stock)
```

```python
# create csv_file
import csv
stock=[['AAPL',datetime.datetime(2014, 6, 20, 0, 0), 90.91],
      ['MSFT',datetime.datetime(2014, 6, 20, 0, 0),  41.68],
       ['FB',datetime.datetime(3014, 6, 20, 0, 0),  64.5],
        ['AAPL',datetime.datetime(2014, 6, 19, 0, 0),  91.86],
        ['MSFT',datetime.datetime(2014, 6, 19, 0, 0),  None],
        ['FB',datetime.datetime(2014, 6, 19, 0, 0),  64.34]
]
f=open('comma_delimited_stock_prices.csv', 'a', encoding='utf8',newline='')
wr=csv.writer(f)
for i in range(len(stock)):
    wr.writerow(stock[i])
```

```python
data:List[StockPrice]=[]

with open('comma_delimited_stock_prices.csv','r',encoding='utf8')as f:
    reader=csv.reader(f)
    for row in reader:
        maybe_stock=try_parse_row(row)
        if maybe_stock is None:
            print(f"skipping invalid row: {row}")
        else:
            data.append(maybe_stock)
print(data)
```

```python
with open('stock.txt','r',encoding='utf8',newline='')as f:
    reader = csv.DictReader(f, delimiter='\t')
    data= [try_parse_row([row['symbol'],row['date'],row['closing_price']])
           for row in reader]
```

```python
from collections import defaultdict

max_prices:Dict[str,float]=defaultdict(lambda: float('-inf'))
print(max_prices)
for sp in data:
    symbol,closing_price=sp.symbol,sp.closing_price
    if closing_price>max_prices[symbol]:
        max_prices[symbol]=closing_price
print(max_prices)
```

```python
# group
prices:Dict[str,List[StockPrice]]=defaultdict(list)

for sp in data:
    prices[sp.symbol].append(sp)
print(prices)
```

```python
prices={symbol:sorted(symbol_prices) for symbol,symbol_prices in prices.items()} # 각겨에 따라 sorting
```

```python
def pct_change(yesterday:StockPrice,today:StockPrice)->float:
    return today.closing_price / yesterday.closing_price -1

class DailyChange(NamedTuple):
    symbol:str
    date:datetime.date
    pct_change:float
    
def day_over_day_changes(prices:List[StockPrice])->List[DailyChange]:
    return [DailyChange(symbol=today.symbol,date=today.date,pct_change=pct_change(yesterday,today))
           for yesterday,today in zip(prices,prices[1:])]

all_changes=[change for symbol_prices in prices.values() for change in day_over_day_changes(symbol_prices)]
```

```python
print(all_changes)
```

```python
max_change=max(all_changes,key=lambda change: change.pct_change)
print(max_change)
```

```python
#척도를 위해 mean, std_dev 사용
from typing import Tuple
from linear_algebra import vector_mean
from statistics import standard_deviation
```

```python
def scale(data: List[Vector])-> Tuple[Vector,Vector]:
    dim=len(data[0])
    means=vector_mean(data)
    stdevs=[standard_deviation([vector[i] for vector in data]) for i in range(dim)]
    return means,stdevs

vectors=[[-3,-1,1],[-1,0,1],[1,1,1]]
means,stdevs=scale(vectors)
print(means,stdevs)
```

```python
def rescale(data:List[Vector])->List[Vector]:
    dim=len(data[0])
    means,stdevs=scale(data)
    
    rescaled=[v[:] for v in data]
    for v in rescaled:
        for i in range(dim):
            if stdevs[i]>0:
                v[i]=(v[i]-means[i])/stdevs[i]
                
    return rescaled

means,stdevs=scale(rescale(vectors))
print(means,stdevs)
```

```python
from linear_algebra import magnitude, dot
```

```python
def direction(w:Vector)->Vector:
    mag=magnitude(w)
    return[w_i/mag for w_i in w]
```

```python
def directional_variance(data:List[Vector],w:Vector)->float:
    w_dir=direction(w)
    return sum(dot(v,w_dir)**2 for v in data)
```

```python
def directional_variance_gradient(data:List[Vector],w:Vector)->float:
    w_dir=direction(w)
    return [sum(2*dot(v,w_dir)*v[i] for v in data) for i in range(len(w))]
```

```python
from gradient_descent import gradient_step
```

```python
def first_principal_component(data:List[Vector], n:int=100, step_size:float=0.1)->Vector:
    guess=[1.0 for _ in data[0]]
    
    with tq.trange(n) as t:
        for _ in t:
            dv=directional_variance(data,guess)
            gradient=directional_variance_gradient(data,guess)
            guess=gradient_step(guess,gradient,step_size)
            t.set_description(f"dv:{dv:3f}")
    
    return direction(guess)
```

```python
from linear_algebra import sclar_multiply
```

```python
def project(v:Vector, w:Vector)->Vector:
    projection_length=dot(v,w)
    return sclar_multiply(projection_length,w)

def remove_projection_from_vector(v:Vector,w:Vector)->Vector:
    return subtract(v,project(v,w))

def remove_projection(data:List[Vector],w:Vector)->List[Vector]:
    return [remove_projection_from_vector(v,w) for v in data]
```

```python
def pca(data:List[Vector],num_components:int)->List[Vector]:
    components:List[Vector]=[]
    for _ in range(num_components):
        component=first_principal_component(data)
        components.append(component)
        data=remove_projection(data,component)
    
    return components
```

```python
def transform_vector(v:Vector,components:List[Vector])->Vector:
    return [dot(v,w) for w in components]

def transform(data:List[Vector],components:List[Vector])->List[Vector]:
    return [transform_vector(v,components) for v in data]
```
