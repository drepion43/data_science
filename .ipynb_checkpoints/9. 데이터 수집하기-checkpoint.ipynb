# 9.2.2 구분자가 있는 파일

```python
# pandas 나 csv 모듈 이용
import csv
import pandas as pd
```

```python
# 헤더가 없는경우
with open('tab_del_stock_prices.txt','r')as f:
    tab_reader=csv.reader(f,delimiter='\t')
    for row in tab_reader:
        date=row[0]
        symbol=row[1]
        closing_price=float(row[2])
        print(date,symbol,closing_price)
```

```python
# 헤더가 있는경우
with open('col_del_stock_price.txt','r')as f: 
    col_reader=csv.DictReader(f,delimiter=':')
    for row in col_reader:
        symbol=row["symbol"]
        date=row["date"]
        price=row["price"]
        print(date,symbol,price)
```

```python
todays_prices={'APPL':90.91, 'MSFT':41.68, 'FB':64.5}

with open('comma_del_stock.price.txt','w')as f:
    csv_writer=csv.writer(f,delimiter=',')
    for stock,price in todays_prices.items():
        csv_writer.writerow([stock,price])
```

# 9.3.2 의회 감시하기

```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://www.house.gov/representatives"
text=requests.get(url).text
soup=BeautifulSoup(text,'html5lib')
print(soup)
```

```python
all_urls=[a['href'] for a in soup('a') if a.has_attr('href')]
print(len(all_urls))
print(all_urls)
```

```python
import re
regex=r"^https?://.*\.house\.gov/?$"
print(re.match(regex,"https://joel.house.gov/"))

good_urls=[url for url in all_urls if re.match(regex,url)]
print(len(good_urls))
print(good_urls)
```

```python
#중복 제거
good_urls=list(set(good_urls))
print(len(good_urls))
```

```python
html=requests.get('https://jayapal.house.gov').text
soup=BeautifulSoup(html,'html5lib')

links={a['href'] for a in soup('a') if 'press release' in a.text.lower()}
print(links)
```

```python
from typing import Dict,Set

press_release:Dict[str,Set[str]]={}

for house_url in good_urls:
    html=requests.get(house_url).text
    soup=BeautifulSoup(html,'html5lib')
    pr_links={a['href'] for a in soup('a') if 'press release' in a.text.lower()}
    print(f"{house_url}: {pr_links}")
    press_release[house_url]=pr_links
```

# 9.4 API 사용하기(JSON,XML)

```python
import json
import requests
```

```python
serialized="""{"title": "Data Science Book",
                "author": "Joel Grus",
                "publicationYear": 2019,
                "topics":["data","science","data science"]}"""
deserialized=json.loads(serialized)
print(deserialized["publicationYear"])
print(deserialized["topics"])
```

```python
github_user="drepion43"
endpoint=f"https://api.github.com/users/{github_user}/repos"
repos=json.loads(requests.get(endpoint).text)
print(repos)
```

```python
from collections import Counter
from dateutil.parser import parse
```

```python
#  create day
dates=[parse(repo["created_at"]) for repo in repos]
month_counts=Counter(date.month for date in dates)
weekday_counts=Counter(date.weekday() for date in dates)
print("month",month_counts)
print("weekday",weekday_counts)
```

```python
# create language
last_5_repo=sorted(repos,key=lambda r: r["pushed_at"], reverse=True)[:5]
last_5_lang=[repo["language"] for repo in last_5_repo]
print(last_5_repo)
print(last_5_lang)
```
