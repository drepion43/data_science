# 12.1 모델

#### 거리를 재는방법
#### 서로 가까운 점들은 유사하다는 가정

```python
from typing import List
from collections import Counter
```

```python
def raw_majority_vote(labels:List[str])->str:
    votes=Counter(labels)
    winner,_=votes.most_common(1)[0]
    return winner
print(raw_majority_vote(['a','b','c','b']))
```

```python
#중복 1등이 있을경우, 단독 1등이 생길때까지 k를 줄이기
def majority_vote(labels:List[str])->str:
    vote=Counter(labels)
    winner,winner_count=vote.most_common(1)[0]
    num_winners=len([count for count in vote.values() if count==winner_count])
    if num_winners==1:
        return winner
    else:
        return majority_vote(labels[:-1])
    
print(majority_vote(['a','b','c','b','a']))
```

```python
#분류기 생성
from typing import NamedTuple
from linear_algebra import Vector,distance
```

```python
class LabeldPoint(NamedTuple):
    point:Vector
    label:str
        
def knn_classify(k:int,labeled_points:List[LabeldPoint],new_point:Vector)->str:
    by_distance=sorted(labeled_points,key=lambda lp:distance(lp.point,new_point))
    k_nearest_labels=[lp.label for lp in by_distance[:k]]
    return majority_vote(k_nearest_labels)
```

# 12.2 Iris Data

```python
import requests
```

```python
data=requests.get("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data")
with open('iris.data','w')as f:
    f.write(data.text)
```

```python
from typing import Dict
import csv
from collections import defaultdict
```

```python
def parse_iris_row(row:List[str])->LabeldPoint:
    measurements=[float(value) for value in row[:-1]]
    label=row[-1].split("-")[1]
    return LabeldPoint(measurements,label)

with open('iris.data')as f:
    reader=csv.reader(f)
    iris_data=[parse_iris_row(row) for row in reader]
    
print(iris_data)
```

```python
points_by_species:Dict[str,List[Vector]]=defaultdict(list)
for iris in iris_data:
    points_by_species[iris.label].append(iris.point)
print(points_by_species)
```

```python
import matplotlib.pyplot as plt
metrics=['sepal length','sepal width','petal length','petal width']
pairs=[(i,j) for i in range(4) for j in range(4) if i< j]
marks=['+','.','x']
fig,ax=plt.subplots(2,3)

for row in range(2):
    for col in range(3):
        i,j=pairs[3*row + col]
        ax[row][col].set_title(f"{metrics[i]} vs {metrics[j]}", fontsize=8)
        ax[row][col].set_xticks([])
        ax[row][col].set_yticks([])
        
        for mark,(species,points) in zip(marks,points_by_species.items()):
            xs=[point[i] for point in points]
            ys=[point[j] for point in points]
            ax[row][col].scatter(xs,ys,marker=mark, label=species)
ax[-1][-1].legend(loc='lower right',prop={'size':6})
plt.show()
```

```python
import random
from machine_learning import split_data
```

```python
random.seed(12)
iris_train,iris_test=split_data(iris_data,0.70)
print(len(iris_train))
print(len(iris_test))
```

```python
# k = 5 로 사용
from typing import Tuple
confusion_matrix:Dict[Tuple[str,str],int]=defaultdict(int)
num_correct=0
```

```python
for iris in iris_data:
    predicted=knn_classify(5,iris_train,iris.point)
    actual=iris.label
    
    if predicted==actual:
        num_correct+=1
        
    confusion_matrix[(predicted,actual)]+=1
    
pct_correct = num_correct/len(iris_test)
print(pct_correct,confusion_matrix)
```

# 12.3 차원의 저주

```python
def random_point(dim:int)->Vector:
    return[random.random() for _ in range(dim)]
print(random_point(10))

def random_distance(dim:int,num_pairs:int)->List[float]:
    return [distance(random_point(dim),random_point(dim)) for _ in range(num_pairs)]
```

```python
import tqdm
dimension=range(1,101)
avg_dis=[]
min_dis=[]
random.seed(0)
for dim in tqdm.tqdm(dimension,desc="Curse of Dimensionality"):
    distances=random_distance(dim,10000)
    avg_dis.append(sum(distances)/10000)
    min_dis.append(min(distances))

min_avg_ration=[min_dist/avg_dist for min_dist,avg_dist in zip(min_dis,avg_dis)]
```

```python
plt.plot(min_dis,'r',label='min')
plt.plot(avg_dis,'b',label='avg')
plt.legend(loc=9)
plt.show()
```

```python
plt.plot(min_avg_ration)
plt.title('min/avg')
plt.show()
```
